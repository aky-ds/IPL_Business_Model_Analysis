```python
def remove_outlires(text):
    q1=ipl_cc['amount_in_crores_2025'].quantile(0.25)
    q3=ipl_cc['amount_in_crores_2025'].quantile(0.75)
    iqr=q3-q1
    upper=q3+1.5*iqr
    lower=q1-1.5*iqr
    if text > upper:
        return upper
    elif text < lower:
        return lower
    else:
        return text
```

```python
ipl_cc['amount_in_crores_2025']=ipl_cc['amount_in_crores_2025'].apply(remove_outlires)
```

```python
def remove_outlires(text):
    q1=ipl_cc['total_deal_value_in_crores'].quantile(0.25)
    q3=ipl_cc['total_deal_value_in_crores'].quantile(0.75)
    iqr=q3-q1
    upper=q3+1.5*iqr
    lower=q1-1.5*iqr
    if text > upper:
        return upper
    elif text < lower:
        return lower
    else:
        return text
```

```python
ipl_cc['total_deal_value_in_crores']=ipl_cc['total_deal_value_in_crores'].apply(remove_outlires)
```

```python
x=ipl_cc.groupby(['partner_sponsor_name','contract_type'])['amount_in_crores_2025'].sum().reset_index().rename(columns={'partner_sponsor_name':'Contributors'})
```

```python
x
```

```python
total_amount=x['amount_in_crores_2025'].sum()
```

```python
x['Percentage Revenue']=round((x['amount_in_crores_2025']/total_amount)*100,2)
```

```python
x.head()
```

```python
contributers_generated_revenue=x.groupby(['Contributors'])['Percentage Revenue'].sum().reset_index()
contributers_generated_revenue.to_csv('contributers_generated_revenue.csv',index=False)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(18, 7))
plt.bar(x['Contributors'], x['Percentage Revenue'], color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel("Contributors", fontsize=14)
plt.ylabel("Percentage Revenue (%)", fontsize=14)
plt.title("Percentage Revenue Contribution by Each Contributor", fontsize=16)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for readability

# Optional: Add percentage labels above bars
for i, v in enumerate(x['Percentage Revenue']):
    plt.text(i, v + 0.5, f"{v:.2f}%", ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```
